```python
#IMPORT LIBRARIES
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#LOAD A SIMPLE DATASET
url="https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df=pd.read_csv(url)
df.head()
```

```python
#TARGET COLUMN IS medv
X=df.drop("medv",axis=1)
y=df["medv"]
```

```python
#TRAIN TEST SPLIT
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#PRINTING SHAPES IS IMPORTANT HABIT
X_train.shape,X_test.shape
```

```python
model=LinearRegression()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)

test_score=r2_score(y_test,y_pred)
```

```python
print("R2 score:",test_score)
```

```python
#Import libraries
import numpy as np
import pandas as pd
```

```python
#Load dataset
df1=pd.read_csv('50_Startups.csv')
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,4]
```

```python
#One hot code encoder dummy variables by considering the categorical data
one_h_e=pd.get_dummies(x['State'],drop_first=True)
```

```python
one_h_e
```

```python
#Drop the old column
x=x.drop('State',axis=1)
```

```python
#Concat the dummy variables
x=pd.concat([x,one_h_e],axis=1)
```

```python
#splitting the dataset into training set and testing set
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=0)
```

```python
#Fitting multiplelinear regression to the training set
regressor.fit(x_train,y_train)
```

```python
#Processing the Test set results
y_pred=regressor.predict(x_test)
```

```python
#Calculating the r2_score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
